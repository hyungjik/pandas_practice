```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

# 한글폰트 사용시 그래프에서 마이너스 폰트 꺠지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
```

```python
df.head()
```

```python
df.plot.scatter(x='a', y='b', s=60, grid=True)   # x와 y 축 데이터 지정이 필수
```

```python
ax = df.plot.scatter(x='a', y='b', color='DarkBlue', label='Group 1')

df.plot.scatter(x='c', y='d', color='DarkGreen', label='Group 2', ax=ax)  # ax를 그 안에 추가
```

```python
df.plot.scatter(x='a', y='b', c='c', s=50)   # c를 추가해 3가지 축의 표현
```

```python
df.plot.scatter(x='a', y='b', s=df['c'] * 200)  # s : c컬럼의 값을 곱해줘서 크기를 변화시킴
```
